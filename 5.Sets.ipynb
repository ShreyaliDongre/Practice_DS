```python
#sets in python
# sets are unordered collections of unique elements 
# they are mutable, meaning you can add or remove elements. 
# creating a set
a = {1, 2, 3, 4, 5}
# adding elements to a set
a.add(6)
# removing elements from a set
a.remove(2)  # raises KeyError if the element is not found
# checking membership in a set
print(3 in a)  # Output: True
# iterating over a set
for element in a:
    print(element)
# converting a list to a set
b = [1, 2, 2, 3, 4]
c = set(b)  # c will be {1, 2, 3, 4}
# sets support mathematical operations like union, intersection, and difference
# union
d = a.union(c)  # d will be {1, 2, 3, 4, 5, 6}
# intersection
e = a.intersection(c)  # e will be {1, 3, 4, 5}
# difference
f = a.difference(c)  # f will be {6}
# update a set with another set
z=a.difference_update(c)  # a will now be {3, 4, 5, 6}
a.update(c)  # a will now include elements from c
# symmetric difference
g = a.symmetric_difference(c)  # g will be {2, 6}
# checking if a set is a subset or superset of another set
print(c.issubset(a))  # Output: False
print(a.issuperset(c))  # Output: True

```

```python
# unique words in a string
def unique_words(s):
    words = s.split() # split the string into words
    unique_set = set(words)  # create a set from the words
    return unique_set  # return the set of unique words
# example usage
s = "hello world hello"
print(unique_words(s))  # Output: {'hello', 'world'}
```
