```python
#BUILT-IN FUNCTIONS
#1. float(): Converts a string or number to a floating-point number.
#2. int(): Converts a string or number to an integer.
#3. str(): Converts a number or other object to a string.
#4. complex(): Converts a number or string to a complex number.
#5. bool(): Converts a value to a Boolean (True or False).      
#6. len(): Returns the length of an object (e.g., string, list).
#7. type(): Returns the type of an object.
#8. range(): Generates a sequence of numbers.
#9. input(): Reads a line of input from the user.
#10. print(): Outputs data to the console.   
#11. max(): Returns the largest item in an iterable or the largest of two or more arguments.
#12. min(): Returns the smallest item in an iterable or the smallest of two or more arguments.
#13. sum(): Returns the sum of all items in an iterable.            
#14. sorted(): Returns a new sorted list from the items in an iterable.
#15. zip(): Combines multiple iterables into a single iterable of tuples.
#16. map(): Applies a function to all items in an iterable and returns a map object.
#17. filter(): Filters items in an iterable based on a function and returns a filter object
#18. all(): Returns True if all elements in an iterable are true.
#19. any(): Returns True if any element in an iterable is true.
#20. abs(): Returns the absolute value of a number.
#21. round(): Rounds a number to a specified number of decimal places.
#22. pow(): Raises a number to a specified power.
#23. divmod(): Returns a tuple containing the quotient and remainder of a division operation.
#24. bin(): Converts an integer to its binary representation as a string.
#25. hex(): Converts an integer to its hexadecimal representation as a string.
#26. oct(): Converts an integer to its octal representation as a string.
#27. chr(): Converts an integer to its corresponding Unicode character.
#28. ord(): Returns the Unicode code point of a character.
#29. id(): Returns the unique identifier of an object.
#30. dir(): Returns a list of valid attributes and methods of an object.
```

```python
#abs()
age=int(input("Enter age: "))
b=abs(age)
print(f"The age is {b} years old.")
```

```python
print(bin(15))
print(hex(15))
print(oct(15))
print(chr(70))  # Unicode character for 'A'
print(ord('l'))  # Unicode code point for 'A'
print(id(age))  # Unique identifier of the age variable
print(dir(age))  # List of attributes and methods of the age variable
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = ['a', 'b', 'c', 'd', 'e']
zipped = zip(list1, list2) # Combine two lists into a list of tuples
#zipped = list(zipped)  # Convert the zip object to a list
for item in zipped:
    print(item)  # Output: (1, 'a'), (2, 'b'), (3, 'c'), (4, 'd'), (5, 'e')
```
