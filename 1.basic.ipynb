```python
#type inference in Python
# Python automatically infers the type of a variable based on the value assigned to it.
# For example, if you assign an integer to a variable, Python treats it as an integer   
# without requiring you to specify the type explicitly.
x = 10  # x is inferred as an integer
y = 3.14  # y is inferred as a float    
z = "Hello"  # z is inferred as a string
# You can check the type of a variable using the `type()` function.
print(type(x))  # Output: <class 'int'>
print(type(y))  # Output: <class 'float'>
print(type(z))  # Output: <class 'str'>
```

```python
#type casting in Python
# Python allows you to convert a variable from one type to another using type casting.  
# You can use built-in functions like `int()`, `float()`, and `str()` to perform type casting.
x = 10.5  # x is a float
y = int(x)  # Convert float to integer
print(y)  # Output: 10
z = str(x)  # Convert float to string
print(z)  # Output: "10.5"  
# You can also convert strings to integers or floats if they represent valid numeric values.
num_str = "42"  
num_int = int(num_str)  # Convert string to integer
print(num_int)  # Output: 42
num_float = float(num_str)  # Convert string to float
print(num_float)  # Output: 42.0
```

```python
#operators in Python
# Python supports various operators, including: 
# - Arithmetic operators: `+`, `-`, `*`, `/`, `%`, `//`, `**`
# - Comparison operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
# - Logical operators: `and`, `or`, `not`
# - Assignment operators: `=`, `+=`, `-=`, `*=`, `/=`, `%=`
# - Bitwise operators: `&`, `|`, `^`, `~`, `<<`, `>>`
# - Membership operators: `in`, `not in`
# - Identity operators: `is`, `is not`
# You can use these operators to perform various operations on variables and values.
# For example:
a = 10
b = 5
sum_result = a + b  # Addition
diff_result = a - b  # Subtraction
multi_result = a * b  # Multiplication
div_result = a / b  # Division
print("Sum:", sum_result)  # Output: Sum: 15
print("Difference:", diff_result)  # Output: Difference: 5
print("Product:", multi_result)  # Output: Product: 50
print("Quotient:", div_result)  # Output: Quotient: 2.0
# Python also supports operator precedence, which determines the order in which operators are evaluated.
# For example, multiplication and division have higher precedence than addition and subtraction.        
# You can use parentheses to change the order of evaluation if needed.
# Python also supports augmented assignment operators, which combine an arithmetic operation with assignment.
# For example:
a += b  # This is equivalent to a = a + b
print("Updated value of a:", a)  # Output: Updated value of a:
# floor division operator (//) performs integer division, discarding the fractional part.
floor_div_result = a // b  # Floor division
print("Floor Division Result:", floor_div_result)  # Output: Floor Division Result: 3
# The exponentiation operator (**) raises a number to the power of another number.
exponent_result = a ** b  # Exponentiation
print("Exponentiation Result:", exponent_result)  # Output: Exponentiation Result: 100000
# Python also supports the modulo operator (%) to find the remainder of a division operation.
mod_result = a % b  # Modulo operation
print("Modulo Result:", mod_result)  # Output: Modulo Result: 0
# Python supports bitwise operators for performing operations on binary representations of integers.
# For example:
bitwise_and = a & b  # Bitwise AND
bitwise_or = a | b  # Bitwise OR
bitwise_xor = a ^ b  # Bitwise XOR
bitwise_not = ~a  # Bitwise NOT
print("Bitwise AND:", bitwise_and)  # Output: Bitwise AND: 0
print("Bitwise OR:", bitwise_or)  # Output: Bitwise OR:
print("Bitwise XOR:", bitwise_xor)  # Output: Bitwise XOR: 15
print("Bitwise NOT:", bitwise_not)  # Output: Bitwise NOT: -11
# Python also supports the left shift (<<) and right shift (>>) operators for shifting bits.
left_shift = a << 1  # Left shift   
print("Left Shift Result:", left_shift)  # Output: Left Shift Result: 20
right_shift = a >> 1  # Right shift
print("Right Shift Result:", right_shift)  # Output: Right Shift Result: 5
# Python supports the `in` and `not in` operators to check for membership in sequences like lists, tuples, and strings. 
# For example:
my_list = [1, 2, 3, 4, 5]
is_in_list = 3 in my_list  # Check if 3 is in the list
is_not_in_list = 6 not in my_list  # Check if 6 is not in the list
print("Is 3 in the list?", is_in_list)  # Output: Is 3 in the list? True
print("Is 6 not in the list?", is_not_in_list)  # Output: Is 6 not in the list? True
# Python also supports the `is` and `is not` operators to check for identity,   
# which checks if two variables refer to the same object in memory.
a = [1, 2, 3]
b = a  # b refers to the same list as a 
is_same_object = a is b  # Check if a and b refer to the same object
is_not_same_object = a is not b  # Check if a and b do not refer
# to the same object
print("Are a and b the same object?", is_same_object)  # Output: Are a and b the same object? True
print("Are a and b not the same object?", is_not_same_object)  # Output: Are a and b not the same object? False
# Python also supports the `and`, `or`, and `not` logical operators for combining boolean expressions.
is_true = True
is_false = False
logical_and = is_true and is_false  # Logical AND
logical_or = is_true or is_false  # Logical OR
logical_not = not is_true  # Logical NOT
print("Logical AND:", logical_and)  # Output: Logical AND: False
print("Logical OR:", logical_or)  # Output: Logical OR: True
print("Logical NOT:", logical_not)  # Output: Logical NOT: False

```
