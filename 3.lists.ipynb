```python
# Creating and working with lists in Python

# Create a list
my_list = [1, 2, 3, 4, 5]

# Append an element
my_list.append(6)

# Remove an element
my_list.remove(3)

# Access elements
first_element = my_list[0]
last_element = my_list[-1]

# Print the list and accessed elements
print("List:", my_list)
print("First element:", first_element)
print("Last element:", last_element)
```

```python
## 1.INTRODUCTION TO LISTS
# Lists are a built-in data structure in Python that allow you to store multiple items in a single variable.
# They are ordered, mutable, and can contain elements of different data types, including other lists.
# Lists are defined using square brackets `[]`, and elements are separated by commas.
#You can add, remove, or modify elements in a list.

## 2.CREATING A LIST
# To create a list, you can use square brackets `[]` and separate the elements with commas.
my_list = [1, 2, 3, 4, 5]
# You can also create an empty list using empty square brackets.
empty_list = []
# You can create a list with different data types, including integers, strings, and even other lists.   
mixed_list = [1, "Hello", 3.14, [2, 4, 6]]  
# You can create a list with a range of numbers using the `range()` function.
number_list = list(range(1, 11))  # Creates a list of numbers from 1 to 10
# You can create a list using the `list()` constructor with an iterable, such as a string or a tuple.
string_list = list("Hello")  # Creates a list of characters in the string "Hello"
# You can create a list with repeated elements using the multiplication operator.
repeated_list = [0] * 5  # Creates a list with five zeros

## 3.ACCESSING ELEMENTS IN A LIST
# You can access elements in a list using their index, which starts at 0 for the first element.
my_list = [10, 20, 30, 40, 50]  
# Accessing the first element
first_element = my_list[0]  
# Accessing the last element
last_element = my_list[-1]
# Accessing a range of elements (slicing)
sliced_elements = my_list[1:4]  # Gets elements from index 1 to 3 (not including index 4)
# You can also use negative indexing to access elements from the end of the list.
negative_indexed_element = my_list[-2]  # Gets the second-to-last element
# You can check if an element exists in a list using the `in` keyword.
element_exists = 20 in my_list  # Returns True if 20 is in the list, otherwise False
# You can find the index of an element using the `index()` method.
element_index = my_list.index(30)  # Returns the index of the first occurrence of 30
# You can count the occurrences of an element using the `count()` method.
element_count = my_list.count(20)  # Returns the number of times 20 appears
# You can iterate through a list using a `for` loop.
for element in my_list:
    print(element)  # Prints each element in the list   
# You can also use list comprehensions to create new lists based on existing ones.
squared_list = [x**2 for x in my_list]  # Creates a new list with the squares of each element


##  4.MODIFYING ELEMENTS IN A LIST
# You can modify elements in a list by accessing them using their index and assigning a new valueto that index.
my_list = [10, 20, 30, 40, 50]
# Modifying the first element
```
