```python
# Example of conditional statements in Python
# if-elif-else:
# This code checks the value of x and prints a message based on its value.
x = 10

if x > 0:
    print("x is positive")
elif x == 0:
    print("x is zero")
else:
    print("x is negative")
```

```python
#Nested conditional statements:
# This code checks if x is positive and then checks if it is even or odd.
x=int(input("Enter a number: ") or 10)  # Default to 10 if no input is given
if x > 0:
    print(x,"is positive")
    if x % 2 == 0:
        print(x,"is even")
    else:
        print(x,"is odd")
elif x == 0:
    print(x,"is zero")
else:
    print(x,"is negative")
```

```python
#Usecase: If a year is a leap year or not:
# This code checks if a given year is a leap year.
year = int(input("Enter a year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, "is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
str="Hello, World!"
#for loop with string:
# This code prints each character in the string.
for char in str:
    print(char)
```

```python
#While loop:
# This code prints numbers from 0 to 4 using a while loop.  
i = 0
while i < 5:
    print(i)
    i += 1  # Increment i by 1 each time through the loop
    
```

```python
#loop control statements:
#break:
# This code breaks out of the loop when i equals 3.
for i in range(5):
    if i == 3:
        break  # Exit the loop when i is 3
    print(i)
```

```python
#continue statement:
# This code skips printing the value when i equals 3.
for i in range(10):
    if i == 3:
        continue  # Skip the rest of the loop when i is 3
    print(i)

```

```python
for i in range(10):
    if i%2!=0:
        continue  # Skip  odd numbers
    print(i)  # Print only even numbers
```

```python
#pass statement:
# This code does nothing when i equals 3, effectively skipping that iteration.
for i in range(5):
    if i == 3:
        pass  # Do nothing when i is 3
    print(i)
```

```python
for i in range(10):
    if i%2!=0:
        continue  # Skip  odd numbers
    print(i)  # Print only even numbers
```

```python
for i in range(3):
    for j in range(2):
        print("i:{i} and j:{j}".format(i=i, j=j))
```

```python
#prime number check:
num = int(input("Enter a number to check if it is prime: "))
is_prime = True
if num < 2:
    is_prime = False    
for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
        is_prime = False
        break
    else:
        is_prime = True
if is_prime:
    print(num, "is a prime number")
```
